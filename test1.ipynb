```python
# generate_requirements.py
import pkg_resources

# List of main packages for the project
packages = [
    "django",
    "djangorestframework",
    "selenium",
]

# Optional: include all currently installed packages
# installed_packages = [p.project_name for p in pkg_resources.working_set]

# Combine your main packages with installed ones if needed
all_packages = packages # list(set(packages + installed_packages))

# Write to requirements.txt
with open("requirements.txt", "w") as f:
    for package in all_packages:
        try:
            version = pkg_resources.get_distribution(package).version
            f.write(f"{package}=={version}\n")
        except pkg_resources.DistributionNotFound:
            f.write(f"{package}\n")

print("requirements.txt has been created!")
```
